```python
import pandas as pd
import numpy as np
df=pd.read_csv("tranzactii_sector2_2024.csv", sep=",")
print(df.dtypes)
```

```python
df.head()
```

```python
df.tail()
```

```python
cols_to_convert = ['S utila', 'S desfasurata', 'S teren']
for col in cols_to_convert:
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)  #se inlocuieste , cu .
    df[col] = pd.to_numeric(df[col], errors='coerce')  #forteaza conversia la float
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum() #suma valorilor lipsa de pe fiecare coloana
```

```python
for col in ['Tip imobil','Tip strada','Zona','Perioada tranzactie']:
    print(f"\n{col}:")
    print(df[col].unique()) #se verifica valorile unice pentru coloane tip text
```

```python
#Cautare valori atipice (outliners)
print(df[['S utila', 'S desfasurata', 'S teren', 'Pret tranzactie LEI']].describe())
```

```python
df['Zona'] = df['Zona'].replace('-', np.nan).str.lower()
df['Zona'] = df['Zona'].replace('0', np.nan)
```

```python
for col in ['Zona']:
    print(f"\n{col}:")
    print(df[col].unique())
```

```python
#Umplem valorile nule din Zona cu cele mai frecvente
moda_zona = df['Zona'].mode()[0]
df['Zona']=df['Zona'].fillna(moda_zona)
```

```python
#Umplem valorile lipsa din coloana "An constructie imobil" cu mediana
mean_an_constructie = df['An constructie imobil'].mean()
df['An constructie imobil']=df['An constructie imobil'].fillna(mean_an_constructie)
```

```python
df.drop(['S desfasurata', 'S teren'], axis=1, inplace=True)
```

```python
#Pastram doar randurile in care S utila nu este 0
df = df[df['S utila'] != 0]

#Apoi verificam daca mai este vreun rand cu 0
print((df['S utila'] == 0).sum())
```

```python
print(df['Perioada tranzactie'].unique())
```

```python
#Eliminare si inlocuire outliers

#S utila
q1_su=df['S utila'].quantile(0.25)
q3_su=df['S utila'].quantile(0.75)
iqr_su=q3_su-q1_su

lower_su=q1_su-1.5*iqr_su
upper_su=q3_su+1.5*iqr_su

out_su=df[(df['S utila']<lower_su) | (df['S utila']>upper_su)]
print(f"S utila - Numar valori atipice: {len(out_su)}")
```

```python
#Inlocuie valori atipice cu mediana
median_su=df['S utila'].median()
df.loc[df['S utila']<lower_su, 'S utila']=median_su
df.loc[df['S utila']>upper_su, 'S utila']=median_su
```

```python
#Pret tranzactie LEI
q1_pret=df['Pret tranzactie LEI'].quantile(0.25)
q3_pret=df['Pret tranzactie LEI'].quantile(0.75)
iqr_pret=q3_pret-q1_pret

lower_pret=q1_pret-1.5*iqr_pret
upper_pret=q3_pret+1.5*iqr_pret

out_pret=df[(df['Pret tranzactie LEI']<lower_pret) | (df['Pret tranzactie LEI']>upper_pret)]
print(f"Pret tranzactie LEI - Numar valori atipice: {len(out_pret)}")
```

```python
#Inlocuie valori atipice cu mediana
median_pret=df['Pret tranzactie LEI'].median()
df.loc[df['Pret tranzactie LEI']<lower_pret, 'Pret tranzactie LEI']=median_pret
df.loc[df['Pret tranzactie LEI']>upper_pret, 'Pret tranzactie LEI']=median_pret
```
